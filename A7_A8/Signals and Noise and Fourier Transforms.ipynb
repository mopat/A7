## Discrete / Fast Fourier Transform

```python
def plotSpectrum(y,Fs):
 """
 Plots a Single-Sided Amplitude Spectrum of y(t)
 http://glowingpython.blogspot.de/2011/08/how-to-plot-frequency-spectrum-with.html
 """
 n = len(y) # length of the signal
 k = arange(n)
 T = n/Fs
 frq = k/T # two sides frequency range
 #frq = frq[0:int(n/2)] # one side frequency range


 Y = fft.fft(y)/n # fft computing and normalization
 #Y = Y[0:int(n/2)]

 plot(frq,abs(Y),'r') # plotting the spectrum
 xlabel('Frequency (Hz)')
 ylabel('Intensity')

Fs = 10.0  # sampling rate of x
y = sin(3*x)

subplot(2,1,1)
plot(x,y)
xlabel('Time')
ylabel('Amplitude')
subplot(2,1,2)
plotSpectrum(y,Fs)
show()
```

```python
subplot(2,1,1)
plot(x[0:50],y[0:50])
plot(x[0:50],(y+n*3)[0:50])
xlabel('Time')
ylabel('Amplitude')
subplot(2,1,2)
plotSpectrum((y+n*3)[0:50],Fs)
show()
```
